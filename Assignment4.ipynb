## 1. Setup
Run these commands in PowerShell to create a venv, install dependencies and initialize the database:

```powershell
python -m venv .venv
.\.venv\Scripts\Activate.ps1
pip install -r requirements.txt
python db_init.py
streamlit run .\streamlit_app.py
```
